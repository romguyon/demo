```python
import numpy as np
import pandas as pd

from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral6
from bokeh.plotting import figure
from bokeh.transform import factor_cmap



#################################             FUNCTIONNALITY             ########################################


# this programs reads a csv file and displays the sums of ITEMS_TOTAL_AMOUNT in the TABLE TBF110 for all months for 

# a determined sender_id 


#################################################################################################################




# Choose your sender_id

senderId = 7597


# Read the csv File

dfsum = pd.read_csv('sumbyyears.csv')

#dfsum.loc[df['SENDER_SITE_LOC_ID'] == some_value]
dfSenderId = dfsum.loc[dfsum['SENDER_SITE_LOC_ID'] == senderId]

print(dfSenderId)

#df1.groupby("year","SENDER_SITE_LOC_ID").count().show()

timeYear= dfSenderId['year'].values
counts = dfSenderId['sum(ITEMS_TOTAL_AMOUNT)'].values

timeYear = [ str(year) for year in timeYear]


source = ColumnDataSource(data=dict(timeYear=timeYear, counts=counts))


# Display in Jupyter notebook

output_notebook()

p = figure(x_range=timeYear, plot_height=350, toolbar_location=None, title="items_sum Counts")
p.vbar(x='timeYear', top='counts', width=0.9, source=source, legend="timeYear", line_color='white', fill_color=factor_cmap('timeYear', palette=Spectral6, factors=timeYear))


# Graph parameters

p.xgrid.grid_line_color = None
#p.y_range.start = 0
#p.y_range.end = 9
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Tooltip : display information when you hover with the mouse
p.add_tools(HoverTool(tooltips=[("sender_id, Date", "7597"), ("counts", "@counts")]))

# Show the graph
show(p)
```
