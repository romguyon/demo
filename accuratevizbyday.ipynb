```python
import numpy as np
import pandas as pd

from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral6
from bokeh.plotting import figure
from bokeh.transform import factor_cmap




#################################             FUNCTIONNALITY             ########################################


# this programs reads a csv file and displays the sums of ITEMS_TOTAL_AMOUNT in the TABLE TBF110 for all days for 

# a determined sender_id for a fixed month and year


#################################################################################################################



# Choose your sender_id

senderId = 15695



# Read the csv file
dfsum = pd.read_csv('sumbyday.csv')


# Select only the data from a specified sender_id

dfSenderId = dfsum.loc[dfsum['SENDER_SITE_LOC_ID'] == senderId]

#print(dftemp)


# Sort data to be displayed in order

dfSenderId = dfSenderId.sort_values(by=['SENDER_SITE_LOC_ID','day'])

print(dfSenderId)

# Set up the graph colors

palettes = ["#c9d9d3", "#718dbf", "#e84d60","#c89e00","#c87500","#c87df2","#c87db5","#c87d62","#847d62","#5b7d62","#103162","#10000a"]


#df1.groupby("year","SENDER_SITE_LOC_ID").count().show()

dayTime= dfSenderId['day'].values
counts = dfSenderId['sum(ITEMS_TOTAL_AMOUNT)'].values

dayTime = [ (str(days)) for days in dayTime]


source = ColumnDataSource(data=dict(dayTime=dayTime, counts=counts))

# Display in Jupyter notebook

output_notebook()

p = figure(x_range=dayTime, plot_height=350, toolbar_location=None, title="items_sum Counts")
p.vbar(x='dayTime', top='counts', width=0.9, source=source, legend="dayTime", line_color='white', fill_color=factor_cmap('dayTime', palette=palettes, factors=dayTime))


# Graph parameters

p.xgrid.grid_line_color = None
#p.y_range.start = 0
#p.y_range.end = 9
p.legend.orientation = "horizontal"
p.legend.location = "top_center"


# Tooltip : display information when you hover with the mouse
p.add_tools(HoverTool(tooltips=[("sender_id, Date", "7597"), ("counts", "@counts")]))



# Show the graph
show(p)
```
