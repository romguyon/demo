```python
import numpy as np
import pandas as pd


#################################             FUNCTIONNALITY             ########################################


# this programs reads a csv file and computes the SUM of TOTAL_ITEMS_AMOUNT in function of CUSTOM_REGIME_CODE and 

# the year


#################################################################################################################


# Read the csv file
df = pd.read_csv('NewTBF110.csv')

senderId = pd.unique(df['SENDER_SITE_LOC_ID'].values)

print(senderId)



df2 = df.groupby(['year', 'SENDER_SITE_LOC_ID','CUSTOMS_REGIME_CODE']).agg({'ITEMS_TOTAL_AMOUNT': 'sum'}).reset_index()



print(df2)


# Remove the duplicates for sender_id

senderId = pd.unique(df2['SENDER_SITE_LOC_ID'].values)


# A table with all the dates which occur one time

dateByYear = pd.unique(df2['year'].values)


customRCode = pd.unique(df2['CUSTOMS_REGIME_CODE'].values)


print(customRCode)

# Set up the table length

lenDateByYear = len(dateByYear)
lenSenderId = len(senderId)
lenCustomRCode = len(customRCode)


# This loop is used to create some rows for sender_id which did not have all the dates and all the custom_regime_code

i =0
while i < lenSenderId:
    for j in range(lenDateByYear):
        for k in range(lenCustomRCode):
            if i+j+k < lenSenderId + lenDateByYear + lenCustomRCode:
                dftemp = df2.loc[df2['SENDER_SITE_LOC_ID'].isin([senderId[i]]) & df2['year'].isin([dateByYear[j]]) & df2['CUSTOMS_REGIME_CODE'].isin([customRCode[k]])]
                #print(senderId[i])
                #print(dateByYear[j])
                if (dftemp.empty == True):
                    df3 = pd.DataFrame(np.array([[dateByYear[j],senderId[i],customRCode[k]]]), columns=['year','SENDER_SITE_LOC_ID','CUSTOMS_REGIME_CODE'])
                    df2 = df2.append(df3, ignore_index=True)

                else :
                    pass


    i = i + 1



df2.drop_duplicates(inplace=True)

df2 = df2.fillna(0)

print(df2)

df2.to_csv('year.csv')
```
