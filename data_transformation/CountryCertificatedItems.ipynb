```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt



# Read the csv file

df = pd.read_csv('TBF115_ITEM_PREF_ORIGIN_DATA_TABLE.csv')


# Replace the Nan Values by 'null'

df['ORIGIN_COUNTRY_CODE_PREF'].fillna('null', inplace=True)


# Condition on origin_preferential (1 means certificated item)

#df = df[df['ORIGIN_PREFERENTIAL'] == 1 ]



# Count the certificated number

certificatedItemsNumber = df['ORIGIN_PREFERENTIAL'].count()





# Computes the percentage of certificated items by country

seriesCertificated = df.groupby('ORIGIN_COUNTRY_CODE_PREF')['ORIGIN_PREFERENTIAL'].count()/certificatedItemsNumber*100



# Transform the pandas series into a dataframe with the speicied columns below

dfcertificated = pd.DataFrame({'ORIGIN_COUNTRY_CODE_PREF':seriesCertificated.index, 'CERTIFICATED_ITEMS':seriesCertificated.values})




# Set up pie parameters
labels = dfcertificated['ORIGIN_COUNTRY_CODE_PREF'].values

sizes = dfcertificated['CERTIFICATED_ITEMS'].values

fig = plt.figure(figsize=[5, 5])
plt.axis('equal')
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True,  radius = 2)


# Plot
plt.show()

```
